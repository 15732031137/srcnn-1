Demonstrate how to use `toolbox` to reproduce SRCNN.

```python
from toolbox.datasets.sr import load_data

# Load data
(x_train, y_train), (x_test, y_test) = load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from toolbox.models import srcnn

# Build model
model = srcnn(x_train.shape[1:])
model.summary()
```

```python
from toolbox.training import train

# Start training
train(model, x_train, y_train, validation_data=(x_test, y_test),
      save_dir='save')
```

```python
# Resume training
train(model, x_train, y_train, validation_data=(x_test, y_test),
      nb_epoch=2, resume=True, save_dir='save')
```
